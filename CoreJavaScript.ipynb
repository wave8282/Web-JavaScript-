프로그래밍 핵심 개념 in JavaScript

---
핵심 개념

  1. 숫자형(Number)
  2. 문자열(String), 문자열 덧셈이 가능
  3. 불 대수(불린형), 참 또는 거짓이 확실한 명제

- typeof 연산자
      - 자료형을 반환하여 준다, 어던 자료형인지 확인


      - 형 변화
      - 자동으로 형 변환
      - 산술 연산자 -> + 기능 주의

- 템플릿 문자열
      - 코드를 가독성 좋게 쉽게 작성 가능
      - S{변수명}

- 자료형
      - null & undefined
      - 값이 없음을 표시, 의도적인 없음 = null
      - 처음부터 없음 = undefined

- 힐당 연산자
- 복합 할당 연산자
- 증가, 감소 연산자

- 함수의 실행 순서
      - 기본적으로 위에서부터 아래로 한줄씩 수행
      - 코드를 실행 하는 중 함수가 실행되면 그 함수를 먼저 실행하고 그 밑으로 내려가면서 순서대로 술행

- return문 제대로 이해하기
      - 값을 돌려주는 Output(결과값)에 해당
      - 함수의 실행을 중단
      - 함수의 내부에서 리턴 키워드가 나오면 함수는 중단됨
      - Dead Code 출력되지 않음

- return과 console.log의 차이
      - console.log는 별 제약없이 반환해줌
      - return은 console.log를 활용하지 않으면 출력되지 않음

- 옵셔녈 파라미터
      - 함수 호출시 기본값을 넣어줄 수 있다
      - 파라미터 = 미리 값 할당
      - 가장 뒤쪽으로 설정해줄 것

- 변수의 scope(범위)
      - 모든 변수는 범위를 벗어나게 되면 오류 발생
      - 로컬변수, 지역변수 : 블록문 내에서만 사용가능
      - 글로벌 변수, 전역변수 : 블록문 밖에서도 사용가능

- 상수(Constant)
      - 절대 변하지 않고 일정한 값
      - 변수let 대신 const를 사용
      - 대문자와 밑주로 구분하여 작성

- if 문(if statament)
      - 조건을 충족하지 않으면 실행하지 않을 수 있다
      - if (조건부분){
          동작부분
        } else{
            동작부분
        }

- else if문
      - 두가지 옵션이 필요한 경우 if문을 한번 더 작성

- switch 문
      - switch (비교할 값){
          case 조건값1;
            동작부분;
            break;
          case 조건값2;
            동작부분;
            break;
          default;
            동작부분;
      }

- for 반복문(loop statement)
      - for (초기화부분; 조건부분; 추가동작부분){
          동작부분
      }
      - ex
      for (let i = 1; i <= 10; i++) {
        console.log(`${i} 코드잇 최고!`);
      }
      - 추가동작부분을 꼭 채울 필요는 없다
      - 초기화 부분에서 생성한 변수는 for문의 로컬변수

- while 반복문
        - while (조건부분){
            동작부분
        }
        - while(i <= 10) {
            console.log("코드잇 반복);
            i++;
        }

- break와 continue
- break문을 만나면 반복이 실행되는 도중 빠져나감
- continue문을 만나면 동작 부분을 한번 건너뜀
