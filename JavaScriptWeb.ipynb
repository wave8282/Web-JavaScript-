JavaScriptWeb

---

- WEB
      - 프론트엔드, 백엔드 etc...
      - 웹이 가진 핵심 요소 : 개발자, PM, 디자이너, 마케터
      - 클라이언트(정보를 제공받는, 요청하는)
      - 서버(서비스를 제공하는)

- 웹 브라우저(훑어보다, 둘러보다), Web Browser
      - 코딩을 하여 서버에 넣어둠
      - 웹 표준(Web Standard)

- 웹 사이트를 만드는 3가지 언어
      - HTML : 뼈대 / 사이트의 화면 틀
      - CSS : 시각적인 옵션들을 설정 / 디자인, 스타일 등
      - JavaScript : 사용자의 행동에 따라 웹 페이지가 반응하는 것

- 크롬 개발자 도구 : 웹을 만드는데 있어 유용

- 서버와 클라이언트가 소통하는 원리
      - IP와 DNS(Domain Name System)
        - IP : 0~256이하의 숫자 4개가 점으로 구분되어있음
        - DNS : 클라이언트가 입력한 도메인에 대한 IP주소를 반환
      - URL : Uniform Resouce Locator
        - ex) https://www.naver.com/index.html
        - 어떤 프로토콜을 사용하는지는 scheme에서 보여줌
      - Query, Anchor

- 정적 웹 페이지(서버 사이드) : 미리 만들어둔 파일을 그래도 제공
- 동적 웹 페이지 : 요청에 따라 서버가 내용을 채워 넣어 매번 새로운 html 제공
  - 클라이언트 사이트/ 서버 사이드

- 데이터만 추가로 받아오기
      - 서버가 필요한 내용만 담아 데이터 제공
      - 요청을 보내면 추가적으로 서버에 요청에 데이터 받아옴
      - 필요한 데이터만 서버에 요청 : 클라이언트 사이드 동적 웹 페이지
      - API : 클라이언트가 서버에 요청할 수 있게 해주는 기능

- 데이터 베이스(DB)
      - 웹 페이지를 요청할 때의 최신 데이터가 포함된 동적 웹 페이지를 만드는 과저응ㄹ 자동화 할수 있음
      - 데이터를 보관해두는 창고, 체계적 관리, 추가와 수저
      - 테이블 : 행과 열로 구성, 세부정보(속성), 엑셀과 비슷
      - 엄청나게 많은 데이터 잘 관리하기 위해 데이터 베이스 사용
      - SQL 언어, 데이터베이스 지식, 유저개개인의 맞춤형
      - 관계형 VS 비관계형
        - 관계형 : 테이블의 형태(행과 열), 테이블간의 관계 SQL 사용 수정,삭제,
        - 비관계형 : 빠르게 개발 시작 가능, 유연한 대응
      
- API(Application Programming Interface)
      - 정해진 방식으로만 소통
      - 개발 생산성 높일 수 있음
      - 웹 API : 클라이언트가 서버에게 HTTP요청을 보내고 서버가 적절한 HTTP응답을 하는

- JSON(JavaScript Object Notation)
      - 클라이언트와 서버가 소통시 데이터를 주고 받는 것
      - 포스트맨(Postman)
    
- WEB 개발 공부 로드맵
      1. 프로튼엔드 개발자(눈에 보이는 영역)
      2. 백엔드 개발자(서버와 소통하는 분야)
