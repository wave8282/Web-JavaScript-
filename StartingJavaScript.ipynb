프로그래밍 시작하기 in JavaScript

---

- 세미클론(;)
      - 문장을 구분하는 방법
      - 문장 구분 반드시 사용

- 코멘트(Comment)
      - 코드를 설명하는 글, 어떤 동작, 어떤 의도
      - // 와 같은 방법을 사용
      - 필요한 만큼만 간결하게 사용

- 자료형(Data Type)
      - 숫자형(Number), 문자형(String), 불린형(Boolean)

- 추상화(Abstraction)
      - 필요한 핵심 특징 추출하여 사용
      - 복잡한 것들 단순화 하여 사용
      - 단순하고 간결하면서 정확하게 표현

- 변수(Variable)
      - let 변수이름;
      - 변수이름에 값 저장 = 등호 사용(할당)
      - 식별자는 문자, 밑줄 혹은 달러기호로 시작
      - 두번째 글자 부터는 숫자 가능
      - 대문자와 소문자는 구별
      - 예약어를 사용하면 안됨
      - 의미없는 이름은 좋지않음, 구체적인 이름 선호
      
- 함수(function)
      - 다양한 명령들을 저장하는 것
      - function 함수이름(){
        명령;
        명령;
        };
      - 함수선언은 정의만 하는 것
      - 함수 호출 -> 함수이름();

- 파라미터/매개변수(Parameter)
      - 함수를 선언하고 호출할 때 소괄호 내부를 적절하게 사용
      - 함수이름(매개변수);
      - 쉼표를 사용하여 여러개의 파라미터 사용
      - function 함수이름(매개변수, 매개변수);


- return 문
      - 함수가 실행된 자리에 값을 돌려주는 것
